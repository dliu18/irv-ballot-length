```python
import matplotlib.pyplot as plt
from scipy import stats
import numpy as np
```

```python
def prob_incorrect_winner(n, p):
    '''
        Each voter votes for the "correct" winner with prob p. If n voters are sampled, what is the prob
        that the correct winner wins?
        
        Equivalent to the CCDF of Bin(n, p)
    '''
    
    threshold = int(n/2)
    return stats.binom.cdf(threshold, n, p)


fig, ax = plt.subplots()

ns = [2**exp for exp in range(8,25)]
ps = [0.5, 0.501, 0.51, 0.6]
for p in ps:
    ax.plot(
        ns,
        [prob_incorrect_winner(n, p) for n in ns],
        label=p
    )
ax.set_xlabel("n")
ax.set_ylabel("Prob of Incorrect Winner")
ax.hlines(0.5, xmin=0, xmax=ns[-1], linestyle="--")
ax.set_xscale("log")

ax.legend()
```
